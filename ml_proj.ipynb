```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
#data = pd.read_csv('crypto_data.csv')

# Group by 'Symbol' and sum the 'Volume'
volume_by_crypto = filtered_data.groupby('symbol')['volume'].sum().reset_index()

# Sort by 'Volume' in descending order
volume_by_crypto_sorted = volume_by_crypto.sort_values(by='volume', ascending=False)

# Select the top 5 cryptocurrencies
top_5_crypto_by_volume = volume_by_crypto_sorted.head(10)

# Display the top 5
print(top_5_crypto_by_volume)

# Plot the top 5 cryptocurrencies by volume
plt.figure(figsize=(10, 6))
sns.barplot(x='volume', y='symbol', data=top_5_crypto_by_volume, palette='viridis')
plt.title('Top 5 Cryptocurrencies by Trading Volume')
plt.xlabel('Total Trading Volume')
plt.ylabel('Cryptocurrency Symbol')
plt.show()
```

inflation_data = inflation_data.resample('D').ffill().reset_index()
Resample the data to a daily ('D') frequency instead of monthly or irregular dates.
ffill() (Forward Fill) fills missing values by carrying the last known value forward.
reset_index() moves the "Date" index back into a normal column.

```python
inflation_data['date'] = pd.to_datetime(inflation_data['date'])
```
